# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&count=100')
url
```

```python
soup = BeautifulSoup(url.content)
name = soup.find("h3",class_='lister-item-header').text.strip().replace('1.\n','').replace('\n','').replace('(1994)','')
name
```

```python
name_ = []
for name in soup.find_all("h3",class_='lister-item-header'):
    name_.append(name.text.strip().replace('\n','').replace('\n','').replace('()',''))
name_  
```

```python
rating = soup.find("div", class_= 'inline-block ratings-imdb-rating')
rating.text.strip()
```

```python
rating_ = []
for r in soup.find_all('div',class_= 'inline-block ratings-imdb-rating'):
    rating_.append(r.text.strip())
rating_
```

```python
year = soup.find("span", class_="lister-item-year text-muted unbold")
year.text
```

```python
year_ = []
for year in soup.find_all(class_="lister-item-year text-muted unbold"):
    year_.append(year.text)
year_
```

```python
df = pd.DataFrame({'Name':name_ ,'Rating':rating_,'Year': year_})
df
```

# Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. b) Top 10 ODI Batsmen along with the records of their team and rating. c) Top 10 ODI bowlers along with the records of their team and rating

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
from bs4 import BeautifulSoup 
import requests
import pandas as pd
url = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
url
```

```python
soup = BeautifulSoup(url.content) #download the page content.

#creating empty lists
team = []
matches = []
points = []
rating = []

#scraping team
for i in soup.find_all('span', class_="u-hide-phablet"):
    team.append(i.text)

#scraping matches played by team 
for i in list(soup.find_all('td',attrs={'table-body__cell u-center-text'}))[::2]:
    matches.append(i.text)

#scraping points
for i in list(soup.find_all('td',attrs={'table-body__cell u-center-text'}))[1::2]:
    points.append(i.text)

#scraping ratings
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    rating.append(i.text)

# Creating data frame 
df = pd.DataFrame({"rank":range(1,11),"Team":team[:10],"Matches":matches[:10] ,"Points":points[:10] ,"Rating":rating[:10]})
df.to_csv('top_10_odi_teams.csv',index = False)
print(df)
```

# Top 10 ODI Batsmen along with the records of their team and rating

```python
from bs4 import BeautifulSoup 
import requests
import pandas as pd
url = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
url
```

```python
soup = BeautifulSoup(url.content)

Rank = range(1,11)
#creating empty lists
name = []
team = []
rating = []

# scraping names
for i in soup.find('div',class_="rankings-block__banner--name"):
    name.append(i)                 
for i in soup.find_all('td',attrs="table-body__cell name")[:9]:
    name.append(i.text.split("\n")[1])

# scraping team name
team1 = (soup.find('div',attrs={"rankings-block__banner--nationality"}).text)
team.append((team1[2:5]).replace('\n',''))
for i in soup.find_all('span',class_="table-body__logo-text")[:9]:
    team.append(i.text.replace('\n',''))
    
#scraping rating
rating.append((soup.find('div',attrs="rankings-block__banner--rating")).text)
for i in soup.find_all('td',class_="table-body__cell u-text-right rating")[:9]:
    rating.append(i.text)
    
    
print(len(name),len(team),len(rating))
```

```python
df = pd.DataFrame({"Name of Batsmen":name,"Team":team,"Rating(Points)":rating}, index = Rank)
df.to_csv('top_10_odi_batsmen.csv')
print(df)
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup = BeautifulSoup(page.content)
#creating empty lists
name = []
cuisine = []
location = []
rating = []
image =[]

# restaurant names
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    name.append(i.text)
    
# cuisine
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split("|")[:][1])

# location
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
# ratings
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
    
# image URL
for i in soup.find_all('img',class_="no-img"):
    image.append(i['data-src'])
    
print(len(name),len(cuisine),len(location),len(rating),len(image))
```

```python
df = pd.DataFrame({'Restaurant name':name,'Cuisine':cuisine,'Location':location,'Ratings':rating,'Image URL':image})
df.to_csv("Restaurants details from dineout.csv")
df
```
